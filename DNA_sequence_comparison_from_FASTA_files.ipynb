function that parses through the file passed in. 

arguments: 
str filepath - name of the file, the path to the file

returns: 
list seq - list of lists of each line of the file, each element of the outer list is a different frog, each element of the inner list is the header and the DNA sequence. 

```python
def readFasta (filepath): 
  """parse FASTA file, store its data"""
  
  seqs = [] # stores file data

  with open (filepath, 'r') as f:

    currentHeader = ''
    currentSeq = ''

    for line in f: 
      # check if header
      if line[0] == '>':
        if currentHeader != '':
          seqs.append([currentHeader, currentSeq])
          currentSeq = ''
        currentHeader = line.strip('\n') 
      else: 
        currentSeq += line.strip('\n')

    seqs.append([currentHeader, currentSeq])

  return seqs 
```

making a variable for each file

```python
mainlandFrogs = readFasta ('/content/drive/MyDrive/frogs_africa.txt')
islandFrogs = readFasta ('/content/drive/MyDrive/frogs_st.txt')
```

creates lists of only the DNA sequence of each frog

```python
islandDNA = []
for frog in islandFrogs: 
  islandDNA.append(frog[1])

mainlandDNA = []
for frog in mainlandFrogs: 
  mainlandDNA.append(frog[1])
```

function that returns the similarity scores of 2 frogs based on how many matches in their DNA sequence

arguments: 
str island - DNA sequence of 1 island frog
str mainland - DNA sequence of 1 mainland frog

returns 
float score/total - gives percentage of matching/total nucleotides compared

```python
def similarityScore(island, mainland): 
  total = 0
  score = 4

  for i in range (len(island)):
    if island[i] == mainland[i] and island[i] != '-' and mainland != '-':
      total += 1
      score += 1
    
    elif island[i] != mainland[i] and island[i] != '-' and mainland[i] != '-':
      total += 1

  return (score/total)
```

compares each Sao Tome frog to all the mainland frogs, titles everything for user comprehension. 

```python
stCounter = 1
for stSeq in islandDNA: 
  mainCounter = 0
  for mainSeq in mainlandDNA:
    simScore = similarityScore(stSeq, mainSeq)
    print ("Sao Tome Frog #" + str(stCounter) + " and " + str(mainlandFrogs[mainCounter][0]) + " similarity score:")
    print (simScore)
    print ("\n")
    mainCounter += 1
  stCounter += 1 
```
